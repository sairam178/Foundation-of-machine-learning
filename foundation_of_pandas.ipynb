DataFrame in pandas

```python
data = {
    "Name":["SAIRAM","sriram","jairam"],
    "Age":[21,22,23],
    "marks":[99,97,8]
}
print(data)
marks = pd.DataFrame(data)
print(marks)
```

```python
marks = pd.DataFrame(
    data,
    index = ["student1","student2","student3"]
)
print(marks)
```

```python
marks.set_index("Name", inplace=True) # if we setting palce = ture it means we changing the orginal objectitself
print(marks)
```
