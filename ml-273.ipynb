Where is numpy


```python
 arr = np.array([1,2,3,4,5])
 print(np.where(arr>2))
```

```python
arr = np.array([1,-5,3,4,5])
```

```python
result = np.where(arr < 0,0,arr)
print(result)
```

```python
result = np.where(arr%2!=0,-1,arr)
print(result)
```

```python
marks =np.array([45,76,29,96])
status= np.where(marks>=50,1,0)
print(status)
```

```python
data = np.array([590,509905,50950,59095])
```

```python
mean_val = np.mean(data)
result = np.where(data>mean_val,mean_val,data)
print(result)
```
