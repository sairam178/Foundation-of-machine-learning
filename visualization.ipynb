```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "name":["aman","riya","sai","ram","sneha","priya"],
    "age":[23,34,44,65,24,55],
    "gender":["male","female","male","male","female","female"],
    "marks":[75,86,100,97,99,54]
}
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["marks"])#as we are analysing only one feature so it is uni varience analysis
plt.title("marks trend")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
plt.bar(df["name"],df["marks"])
plt.title("marks trend")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
plt.hist(df["age"],bins=5)
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```

```python
plt.scatter(df["age"],df["marks"])
plt.title("age vs marks")
plt.xlabel("age")
plt.ylabel("marks")
plt.show()
```

```python
gender_counts = df["gender"].value_counts()
plt.pie(gender_counts,labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

customising plots

example with colors and grid

```python
plt.bar(df["name"],df["marks"],color="red")
plt.title("marks trend")
plt.xlabel("students")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(df["age"])
plt.title("age distribution")

plt.subplot(1,2,2)
plt.bar(df["name"],df["marks"])
plt.title("marks of students")
plt.show()
```
