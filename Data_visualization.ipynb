```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "name":["aman","riya","sai","ram","sneha","priya"],
    "age":[23,34,44,65,24,55],
    "gender":["male","female","male","male","female","female"],
    "marks":[75,86,100,97,99,54]
}
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["marks"])#as we are analysing only one feature so it is uni varience analysis
plt.title("marks trend")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
plt.bar(df["name"],df["marks"])
plt.title("marks trend")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
plt.hist(df["age"],bins=5)
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```

```python
plt.scatter(df["age"],df["marks"])
plt.title("age vs marks")
plt.xlabel("age")
plt.ylabel("marks")
plt.show()
```

```python
gender_counts = df["gender"].value_counts()
plt.pie(gender_counts,labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

customising plots

example with colors and grid

```python
plt.bar(df["name"],df["marks"],color="red")
plt.title("marks trend")
plt.xlabel("students")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(df["age"])
plt.title("age distribution")

plt.subplot(1,2,2)
plt.bar(df["name"],df["marks"])
plt.title("marks of students")
plt.show()
```

```python
data = {
    "monthy_income" :[15000,18000,20000,22000,17000,19000,21000,23000,16000,20000,18000,22000,21000,24000,17000,19000,20000,23000,18000,21000],
    "online_order":[3,4,5,6,4,5,6,7,3,5,4,6,5,7,4,5,6,7,4,5],
    "spotify":[1,2,3,4,2,3,4,5,1,3,2,4,3,5,2,3,4,5,2,3],
    "shop_hours":[2,3,4,5,3,4,5,6,2,4,3,5,4,6,3,4,5,6,3,4],
    "money_spent":[4000,4500,5000,6000,4200,4800,5500,6500,4100,5000,4600,6000,5300,6800,4300,4900,5200,6700,4500,6400]

}
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
correlation_matrix = df.corr(numeric_only=True)

# Display the correlation matrix
print("Correlation Matrix:")
display(correlation_matrix)
```

```python
import matplotlib.pyplot as plt

# Plot 1: Histogram of Money Spent
plt.figure(figsize=(8, 5))
plt.hist(df["money_spent"], bins=10, edgecolor='black')
plt.title("Distribution of Money Spent")
plt.xlabel("Money Spent ($)")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Plot 2: Scatter plot of Monthly Income vs. Money Spent
plt.figure(figsize=(8, 5))
plt.scatter(df["monthy_income"], df["money_spent"], alpha=0.7)
plt.title("Monthly Income vs. Money Spent")
plt.xlabel("Monthly Income ($)")
plt.ylabel("Money Spent ($)")
plt.grid(True)
plt.show()
```

```python
# Plot 3: Bar plot of Online Order Counts
online_order_counts = df["online_order"].value_counts().sort_index()
plt.figure(figsize=(8, 5))
online_order_counts.plot(kind='bar', color='green')
plt.title("Distribution of Online Order Counts")
plt.xlabel("Number of Online Orders")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
