```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "name":["aman","riya","sai","ram","sneha","priya"],
    "age":[23,34,44,65,24,55],
    "gender":["male","female","male","male","female","female"],
    "marks":[75,86,100,97,99,54]
}
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["marks"])#as we are analysing only one feature so it is uni varience analysis
plt.title("marks trend")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
plt.bar(df["name"],df["marks"])
plt.title("marks trend")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
plt.hist(df["age"],bins=5)
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```

```python
plt.scatter(df["age"],df["marks"])
plt.title("age vs marks")
plt.xlabel("age")
plt.ylabel("marks")
plt.show()
```

```python
gender_counts = df["gender"].value_counts()
plt.pie(gender_counts,labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

Count Plot(categorical Data)

Scatter Plot(Relationship between two numerical variables)

```python
sns.scatterplot(x="age",y="fare",data=df)
plt.title("Age vs Fare")
plt.show()
```

Violin Plot (category vs Numerical)

```python
sns.violinplot(x="sex",y="age",data=df)
plt.title("Sex vs Age")
plt.show()
#
```

correlation

```python
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

using IQR inter quartile range  Detecting anomalies /outliers

```python
q1 = df["fare"].quantile(0.25)
q3 = df["fare"].quantile(0.75)
iqr = q3-q1
lower_bound = q1-1.5*iqr
upper_bound = q3+1.5*iqr
outliers = df[(df["fare"]<lower_bound) | (df["fare"]>upper_bound)]
print("number of outliers",len(outliers))
```

detecting patterns and trends
example:survival vs passengers class

```python
sns.countplot(x="class", hue="survived",data=df)
plt.title("survival by passengers class")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris_df = sns.load_dataset('iris')
print("First 5 rows of the Iris dataset:")
display(iris_df.head(30))
```

### Dataset Information

```python
print("Dataset Information:")
iris_df.info()
```

### Descriptive Statistics

```python
print("Descriptive Statistics:")
display(iris_df.describe())
```

### Check for Missing Values

```python
print("Missing Values:")
display(iris_df.isnull().sum())
```

### Data Visualization

#### Histograms for Feature Distributions

```python
iris_df.hist(figsize=(10, 8), bins=15)
plt.suptitle("Distribution of Iris Features")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### Box Plots for Feature Distributions by Species

```python
plt.figure(figsize=(12, 10))
for i, column in enumerate(iris_df.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='species', y=column, data=iris_df)
    plt.title(f'Box plot of {column} by Species')
plt.tight_layout()
plt.show()
```

#### Pair Plot for Relationships Between Features

```python
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.suptitle("Pair Plot of Iris Features by Species", y=1.02)
plt.show()
```

#### Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
correlation_matrix = iris_df.drop('species', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Iris Features")
plt.show()
```
