```python
import pandas as pd
```

```python
data = {
    "name": ["Aman","Saurav","karan","shubham","abhishek",None,"Sammer"],
    "age": [22,23,None,22,25,24,26],
    "marks":[85,None,88,90,56,72,None],
    "subject":["math","physics","math","chemistry","math","physics","chemistry"]

}
df = pd.DataFrame(data)
print(df)
```

```python
print(df.describe())
```

```python
print(df.dtypes)
```

```python
print(df.shape)
```

```python
print(df[["name","marks"]])
```

```python
print(df["marks"]>80)
```

```python
df[(df["age"]>=22) & (df["marks"]>80) ]
```

```python
df["subject"].value_counts()
```

```python
df.sort_values("marks",ascending=False)
```

```python
top = df["marks"].nlargest(2)
print(top)
```

```python
lowest_mark = df['marks'].min()
lowest_marks_students = df[df['marks'] == lowest_mark]
print(lowest_marks_students)
```

```python
print('Missing values before handling:')
print(df.isnull().sum())
```

```python
cleaned_df = df.dropna()
print('DataFrame after removing rows with any missing values:')
print(cleaned_df)
print('\nMissing values in cleaned_df:')
print(cleaned_df.isnull().sum())
```

```python
missing_marks_rows = df[df['marks'].isnull()]
print('Rows with missing marks:')
print(missing_marks_rows)
```

```python
print('Shape of original DataFrame (df):', df.shape)
print('Shape of cleaned DataFrame (cleaned_df):', cleaned_df.shape)
```
