```python
marks = np.array([[78,85,90],[45,60,55],[88,90,95],[30,40,45],[65,70,75]])
print(marks)
```

```python
print(marks.shape)
print(marks.size)
print(marks.ndim)
print(marks.dtype)
```

```python
print(marks[1:2])
```

```python
print(marks[:,2])
```

```python
print(marks[0,1])
```

```python
print(np.mean(marks,axis =0))
print(np.mean(marks,axis =1))
```

```python
print(marks>=80)
```

```python
marks +=5
print(marks)
```

```python
mark = np.where(marks>=50,"1","0")
print(mark)
```

```python
mark = np.where(marks<=40,0,marks)
print(mark)
```

```python
avg = np.mean(marks,axis=1)
print(avg)
```

```python
Grades = np.where(avg>=85,'A',np.where(avg>=70,'B',np.where(avg>=50,'C','0')))
print(Grades)
```

```python
backup = marks.copy()
marks[0,0]=0
print(backup)
print(marks)
```
