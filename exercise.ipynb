```python
import numpy as np
electricity = np.random.randint(120,650,size=(4,7))
print(electricity)
print(electricity.shape)
print(electricity.size)
print(electricity.ndim)
print(electricity.dtype)
average_usage_per_hostel = np.mean(electricity, axis=1)
print("Average electricity usage per hostel:")
print(average_usage_per_hostel)
```

```python
maximum_usage_per_hostel = np.max(electricity, axis=1)
print("Maximum electricity usage per hostel:")
print(maximum_usage_per_hostel)
```

```python
highest_avg_usage_hostel_index = np.argmax(average_usage_per_hostel)
highest_avg_usage_hostel_number = highest_avg_usage_hostel_index + 1
highest_avg_usage_value = average_usage_per_hostel[highest_avg_usage_hostel_index]

print(f"Hostel with the highest average electricity usage: Hostel {highest_avg_usage_hostel_number}")
print(f"Average usage for this hostel: {highest_avg_usage_value:.2f}")
```
